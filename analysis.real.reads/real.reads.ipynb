```python
import pandas as pd
data=pd.read_csv('../Table1.csv')
data['age']=2019-data['Year_of_publication']+1
data['citation_per_year']=data['Number_of_citations']/data['age']


def f(x):
    if x['local'] == 'SW': return 'SW' 
    elif x['local'] == 'hamming distance': return 'Hamming distance'
    elif x['local'] == 'NW': return 'NW'
    elif x['local'] == 'Non-DP heuristic': return 'non-DP heuristic'
    elif x['local'] == 'SW & NW': return 'SW & NW'
    elif x['local'] == 'banded Myres bit vector': return 'banded Myres bit vector'
    elif x['local'] == 'Semi-global': return 'semi-global'
    elif x['local'] == 'Local': return 'SW'
    else: return 'other'

data['local2'] = data.apply(f, axis=1)
data.head()



data.head(3)
```

```python
import pandas as pd
df=pd.read_csv('compare.tools.complete.csv')
df['Aligner'].replace('gsnap', 'GSNAP',inplace=True)
df['Aligner'].replace('hisat2', 'HISAT2',inplace=True)
df['Aligner'].replace('last', 'LAST',inplace=True)
df['Aligner'].replace('lordfast', 'lordFAST',inplace=True)
df['Aligner'].replace('ngm', 'NGMLR',inplace=True)
df['Aligner'].replace('rmap', 'RMAP',inplace=True)
df['Aligner'].replace('smalt', 'Smalt',inplace=True)
df['Aligner'].replace('snap', 'SNAP',inplace=True)
df['Aligner'].replace('soapsplice', 'SOAPSplice',inplace=True)
df['Aligner'].replace('subread', 'Subread',inplace=True)
df['Aligner'].replace('tophat2', 'TopHat2',inplace=True)








```

```python
#adopted from here http://holoviews.org/user_guide/Customizing_Plots.html
import holoviews as hv
from holoviews import opts, dim
import numpy as np


overlay = hv.NdOverlay({group: hv.Scatter(([group]*100, np.random.randn(100)*(5-i)-i))
                        for i, group in enumerate(['A', 'B', 'C', 'D', 'E'])})

errorbars = hv.ErrorBars([(k, el.reduce(function=np.mean), el.reduce(function=np.std))
                          for k, el in overlay.items()])


(errorbars * overlay ).opts(
    opts.ErrorBars(line_width=5), opts.Scatter(jitter=0.2, alpha=0.5, size=6, height=400, width=600))


```
